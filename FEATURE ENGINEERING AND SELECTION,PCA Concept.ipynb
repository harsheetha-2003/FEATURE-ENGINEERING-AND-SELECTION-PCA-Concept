### =>> Insight of what is slope of peak exercise ST segment -

#### The slope of the peak exercise ST segment on an ECG shows how the heart's electrical activity changes during intense exercise.
#### A rising slope is usually good, while a flat or downward slope may indicate heart issues.

```python
slope_target=df.groupby(['slope','target']).size().reset_index().rename(columns={0:'slope_count'})
```

```python
sns.barplot(x='target',y='slope_count',hue='slope',data=slope_target)
#plt.legend(['upslope','flat'],loc='upper center');
#plt.legend(["upslope","flat","downslope"]);
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df)
#it will scale down the values by taking standard deviation as 1 and mean as 0 
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

```python
from sklearn .decomposition import PCA
```

```python
pca=PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
x_pca
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=df['target'])
plt.xlabel('first principle component')
plt.ylabel('second principle component')

```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c=df['target'])
ax.set_xlabel('first principle component')
ax.set_ylabel('second principle component')
ax.set_zlabel('third principle component')
ax.set_title('3D Scatter Plot of the First Three Principal Components')
plt.show()
```
